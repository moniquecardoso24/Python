## Passo 1 - Importar bibliotecas


```python
# Importar bibliotecas 

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Passo 2 - Importar arquivo

```python
# Ler o arquivo

museum_data = pd.read_csv("museum_visitors.csv", index_col="Date", parse_dates=True)

# Printar

display(museum_data.tail())
```

## Passo 3 - Responder as perguntas

```python
# Visitantes de Jul/2018 do Chinese American Museum

print(museum_data.tail()['Chinese American Museum'].loc['2018-07-01'])
```

```python
# Quantidade que o museu Adobe recebeu a mais que o Firehouse

quantidade = museum_data['Avila Adobe'] - museum_data['Firehouse Museum']

print(quantidade.tail().iloc[3])
```

## Passo 4 - Plotar gráfico 

Criar um gráfico de linhas que mostre como o número de visitantes de como cada museu evoluiu ao longo do tempo. A figura deve ter quatro linhas (uma para cada museu).

```python
# Gráfico de linhas 

# Mostrar o número de visitantes de cada museu ao longo dos anos(da tabela)

sns.lineplot(data=museum_data)
```

## Passo 5 - Plotar gráfico para museu Avila Adobe

```python
# Gráfico de linhas

# Criar um gráfico de linhas que mostre como o número de visitantes do Avila Adobe evoluiu ao longo dos anos

sns.lineplot(data=museum_data['Avila Adobe'] , label='Abila Adobe')
```
