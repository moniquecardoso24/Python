## Lists
To create a list use square brackets `[]`, and use commas to separate the values.

```python
# Create a list
my_list1 = ["Python", "is", "fun!"]

# Print the list
print(my_list1)
```

To slice multiple values, use `:` to specify the start and stop (not included) of your slice. 

```python
# Create a list
#0-> Python  -3  
#1-> is      -2
#2 -> fun!   -1

my_list = ["Python", "is", "fun!"]

# Print the first element of my_list
print(my_list[0])  

# Print the last element of my_list
print(my_list[-2])

# Print the first two elements of my_list
print(my_list[0:2])

#More exemples
print(my_list[0:1])  #Start in 0 ends in 1 is Python

my_list2 = ['Good','Inspiration','Positive','amazing','nice']
print(my_list2[1:4])

print(my_list2[0:2])

print(my_list2[0:3])
```

It's possible to replace elements of a list by slicing and assigning a new value.

```python
# Create a list
numbers = [1, 2, 3, 4, 5]

# Change the last value
numbers[-1] = 6   #change 5 to 6

# Print the list
print(numbers)
```

It's possible to extend a list using `+`.

```python
# Create a list
numbers = [1, 2, 3, 4, 5]

# Add additional values
more_numbers = numbers + [6, 7, 8, 9, 10, 11,12]
more_numbers2 = more_numbers + [13,14,15,16]

# Print the list
print(more_numbers2)
```

Use `del()` to delete elements of a list.

```python
# Create a list
numbers = [1, 2, 3, 4, 5]

numbers2 = [6,7,8,9,10,11,12,13]

# Delete the fourth and fifth values in the list
del(numbers[3:])  #delete last 2 elements of a list
del(numbers2[4:]) #delete fourth element until last element

# Print the list
print(numbers)

print(numbers2)
```

## String and list methods
There're a variety of methods we can use on strings, such as capitalization and counting the number of characters. 
(Upper) -> uppercase
(Lower) -> lowercase
(count) -> numbers of a's in this case.

```python
# Define a string variable
x = "data science"

# Convert the string to uppercase
print(x.upper())  # DATA SCIENCE

# Convert the string to lowercase
print(x.lower())  # data science

# Count the number of 'a's
print(x.count("a"))  # 2 a's in DAtA Science
```

Lists also have a number of methods to assist with calculations, searching for values, and adjusting the list.

```python
# Define a list
colors = ["red", "blue", "red", "red", "blue", "green", "purple"]

# Print out the number of times 'red' appears
print(colors.count("red"))  #count the numbers of red

# Print out the index of the first 'red'
print(colors.index("red"))

# Add another element to the list
colors.append("yellow")  #add yellow in this list
print(colors)

# Reverse the list
colors.reverse()  #reverse the elements of list. Yellow was the last element, but now it's the first element
print(colors)
```

Create a NumPy array from a list using `np.array()`.

```python
# Import NumPy as np
import numpy as np
```

```python
# Create a list
numbers = [1, 2, 3, 4, 5, 6, 7]

# Convert the list to an array
numbers_array = np.array(numbers)

# Print the type of numbers_array
print(type(numbers))  #list
print(type(numbers_array))  #numpy.ndarray
```

You can perform mathematical operations on a NumPy array as you would with a value.

```python
# Create an array
numbers = np.array([1, 2, 3, 4, 5])

# Multiply the array by 10
numbers_multiplied = numbers * 10

# Print the array
print(numbers_multiplied)
```

We can subset an array in the same way you would with a list, using square brackets `[]` and colons `:`.

```python
# Create an array
sales = np.array([110, 260, 50, 390, 30])

# Subset the array
first_sales = sales[0:2]

# Print the subset
print(first_sales)
```

We can create a 2D NumPy array by creating a list of lists, and then converting it to an array using `np.array()`.

```python
# Create a list of lists
sales = [[1, 110],
         [2, 260],
         [3, 50]
        ]

# Convert the list to a NumPy array
sales_array = np.array(sales)

# Print the array
print(sales_array)
```

To subset a 2D NumPy array, we use square brackets `[]`, but specify both the row and column index, separated by a comma (`,`).

```python
# Create a list of lists
monthly_sales = [[1, 190],
         [2, 210],
         [3, 185]
        ]

# Convert the list to a NumPy array
sales = np.array(monthly_sales)

# Print 
print(monthly_sales)
```

We can use functions such as `np.mean()`, `np.median()`, and `np.std()` to generate summary statistics of NumPy arrays.

```python
# Create a NumPy array of sales
sales = np.array([190, 210, 185, 235, 170, 270, 240, 225, 260])

# Calculate the average
print(np.mean(sales))   # mean

# Calculate the median
print(np.median(sales))  # median

# Calculate the standard deviation
print(np.std(sales))
```

We can also use `np.corrcoef()` to calculate the correlation between two NumPy arrays.

```python
import numpy as np

# Create a NumPy array of sales
array1 = np.array([0, 1, 2, 3 ,4]) 

# Create a NumPy array of marketing expenses
array2 = np.array([5, 6, 7, 8, 9])

# Calculate the correlation between the two arrays
result = np.corrcoef(array1,array2)

print(result)
```
