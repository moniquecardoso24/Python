```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Initializing list
valores = [1, 2, 3, 4, 5]
quadrados = [valores ** 2 for valores in valores]
cubos = [valores ** 3 for valores in valores]

# Converting list to array
quadrados2 = np.array(quadrados)
cubos2 = np.array(cubos)
```

```python
# Showing the graphic

plt.plot(quadrados2)
plt.plot(cubos2)
```

```python
# Showing the graphic

plt.plot(quadrados2, color = 'red', linewidth = '4', label = 'quadrados')
plt.plot(cubos2, color = 'green', linewidth = '4', label = 'cubos')

plt.legend()  #show legends

# Label 
plt.title("Valores elevado ao quadrado e cubo")
plt.xlabel("Valores", fontsize=15)
plt.ylabel("Quadrados", fontsize=15)
```

```python
# Initializing list

bandas = ['Metallica', 'Slayer', 'Megadeth']
albuns = [10, 11, 15]
```

```python
# Bar graphic

#x-> bandas
#y -> albuns

plt.bar(bandas, albuns, color='green')
plt.title("Albuns de Estudio")
plt.xlabel("Banda", fontsize=14)
plt.ylabel("Quantidade de albuns", fontsize=14)
```
